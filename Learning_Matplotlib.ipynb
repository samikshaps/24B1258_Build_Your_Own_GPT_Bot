```python
#importing
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#simple plotting
x_1=np.linspace(0,5,10)       #generates an array of 10 evenly spaced values between 0 and 5
y_1=x_1**2
plt.plot(x_1,y_1)
plt.title('days squared chart')
plt.xlabel('days')
plt.ylabel('days squared')
plt.show()
```

```python
#printing multiple plots
plt.subplot(1,2,1) #plt.subplot(nrows, ncols, index) note that index starrt from 1 and not 0; index increases row-wise(row1col1=1, row1,col2=2, row2col1=3 etc); giving index activates that plot
plt.plot(x_1,y_1,'r')
plt.subplot(1,2,2)
plt.plot(x_1,y_1,'b')
plt.show()
```

```python
#using figure objects
fig_1=plt.figure(figsize=(5,4), dpi=100)  #figsize(width,height), dpi = dots per inch (resolution)
axes_1=fig_1.add_axes([0.1,0.1,0.9,0.9])  #[x0, y0, width, height] these values lie bw 0 and 1
axes_1.set_xlabel('days')
axes_1.set_ylabel('days squared')
axes_1.set_title('days squared chart')
axes_1.plot(x_1,y_1,label='x/x^2')
axes_1.plot(y_1,x_1,label='x^2/x')
axes_1.legend(loc=0) #loc=0 places the legend according to available space; upper right 1; upper left 2; lower left 3; lower right 4
axes_2=fig_1.add_axes([0.45,0.45,0.4,0.3])
axes_2.set_xlabel('days')
axes_2.set_ylabel('days squared')
axes_2.set_title('days squared chart')
axes_2.plot(x_1,y_1,'r')
axes_2.text(0,40,'practice')
```

```python
#subplots
fig_2,axes_2=plt.subplots(figsize=(8,4), nrows=1,ncols=3)
plt.tight_layout() #for spacing
axes_2[1].set_xlabel('x')
axes_2[1].set_ylabel('x^2')
axes_2[1].set_title('plot2')
axes_2[1].plot(x_1,y_1)  #here indexing starts from 0 only
```

```python
#appearance options
fig_3=plt.figure(figsize=(6,4))
axes_3=fig_3.add_axes([0,0,1,1])
axes_3.plot(x_1,y_1, color='navy', alpha=.75, lw=2, ls='-.', marker='o', markersize=7, markerfacecolor='y', markeredgecolor='y', markeredgewidth=4)
#to zoom in
axes_3.set_xlim([0,3])
axes_3.set_ylim([0,8])
#adding background grids and colours
axes_3.grid(True, color='0.6',dashes=(5,2,1,2))  #0.6 stands for gray - 60%black    # dashes(5,2,1,2) - 5 pts , 2 spaces, 1 pt, 2 spaces
axes_3.set_facecolor('m')
```

```python
#saving outputs
fig_3.savefig('plot3.png')
```

```python
#histograms
arr_1=np.random.randint(1,7,6000)
arr_2=np.random.randint(1,7,6000)
arr_3=arr_1+arr_2
#bins - specifies the number of bins (or intervals) to divide the data into; here the sum of die rolls ranges from 2 to 12 which is 11 values
#density=True:histogram is normalized so that the area under the histogram sums to 1;
#converts the counts in each bin into probabilities or density values, making the histogram represent the probability distribution of the sums of the dice rolls.
#stacked=True:controls whether the bars of the histogram are stacked on top of each other or not
plt.hist(arr_3, bins=11, stacked=True, density=True, cumulative=True, color='orange')
#cumulative=True comment and uncomment it to see what it does
#orientation='horizontal'
#histtype='step' generates a line plot
```

```python
#bar charts
x=['nuclear', 'hydro','coal','gas','solar','wind','other']
per_1=[71,10,3,7,2,4,3]
variance=[8,3,1,3,1,2,1]
plt.bar(x,per_1,color='purple',yerr=variance) #the black vertical lines show the variance
#plt.barh will plot horizontally
```

```python
#bar charts
m_eng=[76,85,86,88,93]
f_eng=[24,15,14,12,7]
spc=np.arange(5)
plt.bar(spc, m_eng, edgecolor='k', label='Male', width=.45)
plt.bar(spc+0.45, f_eng, edgecolor='k', label='Female', width=.45)
plt.xticks(spc+0.45/2, ('aero','chem','civil','elec','mech'))
```

```python
#bar charts
t_type=['kind','elem','sec','spec']
m_teach=np.array([2,20,44,14])
f_teach=np.array([98,80,56,86])
ind=[x for x, _ in enumerate(t_type)]
#enumerate() returns pairs of (index, value) for each element in the t_type
#[x for x, _ in enumerate(t_type)] - list comprehension; extracts only x from the index value pairs;
#underscore is a common placeholder variable name in Python used when the value isn't needed
plt.bar(ind, m_teach, width=.45, label='Male', bottom=f_teach)
plt.bar(ind, f_teach, width=.45, label='Female')
plt.legend(loc='lower right')
```

```python
#pie charts
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]
myexplode = [0.2, 0, 0, 0] #apples will be exploded by0.2 others remain at centre
plt.pie(y, labels = mylabels, explode = myexplode, shadow=True, colors=mycolors)
plt.legend(title = "Four Fruits:")
plt.show()
```
